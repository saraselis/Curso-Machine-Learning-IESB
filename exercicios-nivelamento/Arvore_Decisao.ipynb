```
import numpy as np
import matplotlib.pyplot as plt
from csv import reader
import pandas as pd
import seaborn as sns
```

```
sns.set(style='ticks', color_codes=True)
```

```
df = sns.load_dataset("iris")
```

```
df.head(5)
```

```
df.shape
```

```
df['species'].value_counts().plot(kind='bar')
```

```
sns.pairplot(df, hue='species', palette='husl')
```

```
X = df.drop('species',axis=1)
y= df['species']
```

```
from sklearn.tree import DecisionTreeClassifier
```

```
clf = DecisionTreeClassifier(max_depth=2) #max_d profundida da arvore, definir ele ate otimizar
```

```
clf.fit(X, y)
```

```
from sklearn.tree import export_graphviz
from graphviz import Source
from subprocess import call
```

```
dot = export_graphviz(clf, out_file="tree.dot", feature_names=df.columns[:-1], class_names=['setosa','versicolor','virginica'], rounded=True, filled=True)
call(['dot', '-Tnpg', 'tree.dot', '-o','tree.png','-Gdpi=600'])
```
